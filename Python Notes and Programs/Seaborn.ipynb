```python
import numpy as np
```

```python
a=np.linspace(0,np.pi*2)
```

```python
b=np.sin(a)
```

```python
plt.plot(a,b,color="r")
plt.title("Sin(x) Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
plt.show()
```

```python
a=np.linspace(0,np.pi*2)
b=np.cos(a)
plt.plot(a,b,color="b")
plt.title("Cos(x) Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
plt.show()
```

```python
a=np.linspace(0,np.pi*2)
b=np.tan(a)
plt.plot(a,b,color="g")
plt.title("Tan(x) Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
plt.show()
```

```python
# data

# Dataframe

# hue

# Variable in data to map plot aspects to different colors.

# palette

# Set of colors for mapping the hue variable

# kind

# Kind of plot for the non-identity relationships. {scatter, reg}

# diag_kind

# Kind of plot for the diagonal subplots. {hist, kde}
```

```python
#Pairgrif
```

```python
g = sb.PairGrid(df)
g.map(plt.scatter,color="orange")
plt.show()
```

```python
g = sb.PairGrid(df)
g.map_diag(plt.hist)#diagonal plot
g.map_offdiag(plt.scatter);#offdiagonal
plt.show()
```

```python
g = sb.PairGrid(df)
g.map_upper(plt.scatter)
g.map_lower(sb.kdeplot, cmap = "Blues_d")
g.map_diag(sb.kdeplot, lw = 3, legend = False);
plt.show()
```
