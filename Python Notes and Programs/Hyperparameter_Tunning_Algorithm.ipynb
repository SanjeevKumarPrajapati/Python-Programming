```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.drop("Species",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop("Id",axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(2,21):
  km=KMeans(n_clusters=i,init="k-means++")
  km.fit(df)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(12,6))
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xticks([i for i in range(2,21)])
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.grid(axis="x")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
ss=[]
no_c=[i for i in range(2,21)]

for i in range(2,21):
  km=KMeans(n_clusters=i,init="k-means++")
  km.fit(df)
  ss.append(silhouette_score(df,km.labels_))
```

```python
plt.figure(figsize=(12,6))
plt.plot(no_c,ss,marker="o")
plt.xticks(no_c)
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.grid(axis="x")
plt.show()
```

```python
kmn=KMeans(n_clusters=3)
kmn.fit(df)
```

```python
df["Predict"]=kmn.predict(df)
```

```python
df
```

```python
sns.pairplot(data=df,hue="Predict")
plt.savefig("kmean.jpg")
plt.show()
```

##Max Voting and Average Voting

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=10000,noise=0.8)
```

```python
df=pd.DataFrame({"X1":x[:,0],"X2":x[:,1],"Y":y})
```

```python
df
```

```python
sns.scatterplot(x="X1",y="X2",data=df,hue="Y")
plt.show()
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
dt.score(X_test,Y_test)*100,dt.score(X_train,Y_train)*100
```

```python
svm=SVC()
svm.fit(X_train,Y_train)
svm.score(X_test,Y_test)*100,svm.score(X_train,Y_train)*100
```

```python
nb=GaussianNB()
nb.fit(X_train,Y_train)
nb.score(X_test,Y_test)*100,nb.score(X_train,Y_train)*100
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vt=VotingClassifier(estimators=[("dt",dt),("svm",svm),("nb",nb)],weights=[5,4,3])
```

```python
vt.fit(X_train,Y_train)
```

```python
vt.score(X_train,Y_train)*100,vt.score(X_test,Y_test)*100
```

```python
pred=pd.DataFrame({"dt":dt.predict(X_test),"svm":svm.predict(X_test),"nb":nb.predict(X_test),"vc":vt.predict(X_test)})
```

```python
sv=SVR()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
vt.fit(x_train,y_train)
```

```python
vt.score(x_test,y_test)*100,vt.score(x_train,y_train)*100
```

##Bagging & Random Forest Classifier

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC(),n_estimators=20)
bg.fit(x_train,y_train)
bg.score(x_test,y_test)*100,bg.score(x_train,y_train)*100
```

```python
sv=SVC()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
rf=RandomForestClassifier(n_estimators=20,criterion="gini",max_depth=6)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)*100,rf.score(x_train,y_train)*100
```
