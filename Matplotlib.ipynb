# Line Plot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.array([10,20,30,40])
```

```python
y=x*2
```

```python
plt.plot(x,y,linestyle=":",color="red",linewidth=5)
plt.title("Line Plot")
plt.xlabel("X->Axis")
plt.ylabel("Y->Axis")
plt.show()
```

# Bar Plot

```python
student={"Rahul":87,"Matt":56,"Sam":27,"Amit":89}
```

```python
names=list(student.keys())
```

```python
values=list(student.values())
```

```python
plt.bar(names,values,color="red")
plt.title("Bar Plot")
plt.xlabel("X->Axis")
plt.ylabel("Y->Axis")
plt.show()
```

# Scatter Plot

```python
x=[10,20,30,34,12,34.78]
```

```python
y=[23,34,45,56,67.78,23.89]
```

```python
plt.scatter(x,y,color="green")
plt.title("Scatter Plot")
plt.xlabel("X->Axis")
plt.ylabel("Y->Axis")
plt.show()
```

# Histogram

```python
data=[1,3,3,3,3,9,9,5,4,3,2,2,5,6,7,8,9]
```

```python
plt.hist(data,color="green")
plt.title("Histogram Plot")
plt.xlabel("X->Axis")
plt.ylabel("Y->Axis")
plt.show()
```

# Box Plot

```python
one=[1,2,3,4,5,6,7,8,9]
```

```python
two=[1,2,3,4,5,4,3,2,1]
```

```python
three=[6,7,8,9,5,4,3,2]
```

```python
data=list([one,two,three])
```

```python
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("X->Axis")
plt.ylabel("Y->Axis")
plt.show()
```

# Violin Plot

```python
plt.violinplot(data)
```

# Pie Chart

```python
fruits=["Apple","Orange","Mango","Guava","Banana"]
```

```python
quantity=[67,78,56,89,45]
```

```python
plt.pie(quantity,labels=fruits,autopct="%0.1f%%",colors=["red","green","brown","pink","orange"],explode=(0,0,0.1,0.1,0))
plt.show()
```
